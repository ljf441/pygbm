```python
from pygbm import GBMSimulator, EulerMaruyama, MilsteinSimulator
import matplotlib.pyplot as plt

# Parameters for GBM
y0 = 1.0
mu = 0.05
sigma = 0.2
T = 1.0
N = 1000

# Initialize simulator
simulator = GBMSimulator(y0, mu, sigma)
simulator2 = EulerMaruyama(y0, mu, sigma)
simulator3 = MilsteinSimulator(y0, mu, sigma)

# Simulate path
t_values, y_values = simulator.simulate_path(T, N)
t_values2, y_values2 = simulator2.simulate_empath(T, N)
t_values3, y_values3= simulator3.simulate_milpath(T, N)

# Plot the simulated path
plt.plot(t_values, y_values, label="GBM Path")
plt.plot(t_values2, y_values2, label="EM Path")
plt.plot(t_values3, y_values3, label="MIL Path")
plt.xlabel("Time")
plt.ylabel("Y(t)")
plt.title("Simulated Geometric Brownian Motion Path")
plt.legend()
plt.show()
```
